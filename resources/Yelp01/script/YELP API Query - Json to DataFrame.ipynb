```python
import os
import pandas as pd
import json
import time
import operator
import ast
```

```python
#go to the dir with all json files
os.chdir("./Json Data")
```

```python
alldata=pd.DataFrame(columns=('alias', 'categories', 'coordinates', 'display_phone', 'distance', 'id',
       'image_url', 'is_closed', 'location', 'name', 'phone', 'price',
       'rating', 'review_count', 'transactions', 'url'))
```

```python
for file in os.listdir("./"):
        print(file)
        try:
            with open(file) as json_file:  
                response = json.load(json_file)
            df = pd.DataFrame.from_dict(response['businesses'])
            alldata=pd.concat([alldata,df], ignore_index=True)
        except Exception:
            pass
```

```python
finaldf=alldata.drop_duplicates(subset=['id'], keep='first', inplace=False).reset_index(drop=True)
```

```python
finaldf.shape
```

```python
#Simplify the category column
def get_category(x):
    y=ast.literal_eval(x)
    z=[y[j]['alias'] for j in range(len(y))]
    return(z)
```

```python
finaldf['categories'] = finaldf['categories'].apply(lambda x: get_category(x))
```

```python
finaldf.head()
```

```python
#filter again because some records are not coffee shop
iscoffeeshop=[('cafes' in cat) or ('coffee' in cat) or ('coffeeroasteries' in cat) or ('themedcafes' in cat) for cat in finaldf.categories.values] 
```

```python
notcoffee=list(map(operator.not_, iscoffeeshop))
```

```python
sum(notcoffee)
```

```python
finaldf.loc[notcoffee]
#these can be donut shops, burger stands, hotdog stands...
```

```python
finaldf = finaldf.loc[iscoffeeshop]
```

```python
#get the lat and long for easy future calculation
finaldf['long']=finaldf['coordinates'].apply(lambda x: ast.literal_eval(x)['longitude'])
finaldf['lat']=finaldf['coordinates'].apply(lambda x: ast.literal_eval(x)['latitude'])
```

```python
#Also remove foodtrucks
notfoodtrucks=[('foodtrucks' not in cat) for cat in finaldf.categories.values]
finaldf=finaldf.loc[notfoodtrucks]
```
